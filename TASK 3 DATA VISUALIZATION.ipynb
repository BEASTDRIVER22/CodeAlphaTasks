# Titanic Dataset - Data Visualization

## Objective

The goal of this task is to transform raw data into meaningful visualizations that help understand survival patterns and support decision-making.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("titanic.csv")
```

```python
sns.barplot(x='Sex', y='Survived', data=df)
plt.title("Survival Rate by Gender")
plt.show()
```

### Insight

Females had a significantly higher survival rate. This suggests that evacuation priority may have been given to women.

```python
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title("Survival Rate by Passenger Class")
plt.show()
```

### Insight

Passengers in 1st class had better survival rates, indicating that socioeconomic status influenced survival chances.

```python
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```

### Insight

Most passengers were young adults. However, age alone does not strongly determine survival.

```python
sns.histplot(df['Fare'], bins=30, kde=True)
plt.title("Fare Distribution")
plt.show()
```

### Insight

Fare distribution is highly skewed, with a small number of passengers paying significantly higher prices.

## Data Story

The analysis shows that survival on the Titanic was not random. Gender and passenger class played a crucial role. Women and higher-class passengers had significantly better chances of survival, suggesting prioritized evacuation and access advantages. Fare also indirectly reflects this trend, reinforcing the role of socioeconomic factors.

## Conclusion

- Gender and class were the most important survival factors
- Higher fare increased survival likelihood
- Visualizations clearly highlight inequality in survival outcomes
